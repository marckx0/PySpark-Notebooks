```
from pyspark.sql import SparkSession
from pyspark import SparkContext

SpSession = SparkSession \
  .builder \
  .appName("DBA - Spark Actions") \
  .getOrCreate()

SpContext = SpSession.sparkContext
```

```
collData = SpContext.parallelize([4,3,8,5,8])
inputPath = "/FileStore/tables/auto_data.csv"
autoData = SpContext.textFile(inputPath)
autoData.cache()
```

```
autoData.take(3)
```

```
suma = collData.reduce(lambda x,y: x+y)
print("suma ",suma )
```

```
short = autoData.reduce(lambda x,y: x if len(x)<len(y) else y)
print("short ",short )
```

```
def getMPG(autoStr):
  if isinstance(autoStr,int):
    return autoStr
  attList = autoStr.split(",")
  if attList[9].isdigit():
    return int(attList[9])
  else:
    return 0
```

```
promedio = autoData.reduce(lambda x,y : getMPG(x)+getMPG(y))/(autoData.count()-1)
print("Promedio: ",promedio)
```
